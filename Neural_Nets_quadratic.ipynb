```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

xs = np.array([-4.0, -3.0, -2.0, -1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
x_squared = np.array([16.0, 9.0, 4.0, 1.0, 0.0, 1.0, 4.0, 9.0, 16.0], dtype=float)
X = np.column_stack((xs, x_squared))

#y = 2x^2 + 3x + 5
ys = 2 * xs**2 + 3 * xs + 5

model = Sequential([Dense(units=1, input_shape=[2])])

model.compile(optimizer='sgd', loss='mean_squared_error')

model.fit(X, ys, epochs=500)

X_test = np.array([[10.0, 100.0]])
print(model.predict(X_test))

print("Here is what I learned: {}", model.layers[0].get_weights())
```
